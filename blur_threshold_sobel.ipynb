# **Binary Threshold


```python
plt.imshow(apple_img_rgb)
```

```python
ret,threshold = cv2.threshold(apple_img_rgb,127,255,cv2.THRESH_BINARY)
print (ret)
print (threshold.shape)
```

```python
plt.imshow(threshold,cmap='gray')
# this does not work as it needs grayscale image
#https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html
The function cv.threshold is used to apply the thresholding. 
The first argument is the source image, which should be a grayscale image. The second argument is the threshold value which is used to classify the pixel values.
```

```python
gray_image=cv2.imread('apple.jpeg',0)
print (gray_image.shape)
#above prints (531, 469) we are reading it as two d array.
# refer this as to why we need cmap='gray'
#https://stackoverflow.com/questions/65286901/plt-imshownp-squeezex-train3-cmap-gray-what-does-this-command-mean
plt.imshow(gray_image,cmap='gray')
```

```python
#in above image, most of data is near black , more than 127.
#only few points seem to be below 127
#when thresholding is applied, those few points are made to 0(white)
#most other points are set to 255 (black)
#background is already white, already 0, so no change to that
```

```python
ret2,threshold2 = cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)
print (ret2)
print (threshold2.shape)
```

```python
plt.imshow(threshold2,cmap='gray')
```

```python
ret3,threshold3= cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY_INV)
print (ret3)
print (threshold3.shape)
plt.imshow(threshold3,cmap='gray')
```

# **Blur image
**3 * 3 kernel filter
**2d conv

```python
import numpy as np
kernel_filter = np.ones(shape=(5,5))/15
kernel_filter
```

```python
apple_img=cv2.imread('apple.jpeg')
apple_img_rgb = cv2.cvtColor(apple_img,cv2.COLOR_BGR2RGB)
plt.imshow(apple_img_rgb)
```

```python
blurred_image = cv2.filter2D(apple_img_rgb,-1,kernel_filter)
# -1 means use same depth as original
#https://docs.opencv.org/4.x/d4/d86/group__imgproc__filter.html#filter_depths
plt.imshow(blurred_image)
```

```python
tiger_img=cv2.imread('new_tiger.jpg')
tiger_img_rgb = cv2.cvtColor(tiger_img,cv2.COLOR_BGR2RGB)
plt.imshow(tiger_img_rgb)
```

```python
blurred_tiger_image = cv2.filter2D(tiger_img_rgb,-1,kernel_filter)
plt.imshow(blurred_tiger_image)
```

# Sobel Edge detection

```python
apple_img=cv2.imread('apple.jpeg')
apple_img_rgb = cv2.cvtColor(apple_img,cv2.COLOR_BGR2RGB)
plt.imshow(apple_img_rgb)
```

```python
sobeled_x = cv2.Sobel(apple_img_rgb,cv2.CV_8U,1,0,ksize=5)
plt.imshow(sobeled_x)
```

```python
sobeled_x = cv2.Sobel(apple_img_rgb,-1,1,0,ksize=5)
plt.imshow(sobeled_x)
```

```python
sobeled_x = cv2.Sobel(apple_img_rgb,cv2.CV_64F,1,0,ksize=5)
plt.imshow(sobeled_x)
```

```python
https://www.quora.com/Why-can-Sobel-operator-be-only-applied-to-grayscale-images
```

```python
gray_apple_img=cv2.imread('apple.jpeg',0)
plt.imshow(gray_apple_img,cmap='gray')
```

```python
sobeled_x = cv2.Sobel(gray_apple_img,cv2.CV_64F,1,0,ksize=3)
plt.imshow(sobeled_x,cmap='gray')
```

```python
sobeled_x = cv2.Sobel(gray_apple_img,cv2.CV_64F,1,0,ksize=9)
plt.imshow(sobeled_x,cmap='gray')
#this looks better
```

```python
sobeled_x = cv2.Sobel(gray_apple_img,-1,1,0,ksize=9)
plt.imshow(sobeled_x,cmap='gray')
```

```python
gray_tiger_img=cv2.imread('new_tiger.jpg',0)
print (gray_tiger_img.shape)
plt.imshow(gray_tiger_img,cmap='gray')
```

```python
sobeled_tiger = cv2.Sobel(gray_tiger_img,cv2.CV_64F,1,0,ksize=29)
#need to use large kernel
plt.imshow(sobeled_tiger,cmap='gray')
```
