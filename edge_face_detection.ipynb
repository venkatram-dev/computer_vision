```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
tiger_img=cv2.imread('new_tiger.jpg')
tiger_img_rgb=cv2.cvtColor(tiger_img,cv2.COLOR_BGR2RGB)
plt.imshow(tiger_img_rgb)
```

# edge detection

## with out blur

```python
edge_tiger_img_rgb=cv2.Canny(image=tiger_img_rgb, threshold1=127 , threshold2=127)
plt.imshow(edge_tiger_img_rgb)
```

```python
edge_tiger_img_rgb=cv2.Canny(image=tiger_img_rgb, threshold1=80 , threshold2=120)
plt.imshow(edge_tiger_img_rgb)
```

```python
med_val = np.median(tiger_img_rgb) 
lower = int(max(0, 0.7* med_val))
upper = int(min(255,1.3 * med_val))
print (lower)
print (upper)
```

```python
edge_tiger_img_rgb=cv2.Canny(image=tiger_img_rgb, threshold1=60 , threshold2=100)
plt.imshow(edge_tiger_img_rgb,cmap='gray')
```

# after blur

```python
blurred_tiger_img = cv2.GaussianBlur(tiger_img_rgb,(5,5),0)
```

```python
edge_tiger_img_rgb=cv2.Canny(image=blurred_tiger_img, threshold1=80 , threshold2=120)
plt.imshow(edge_tiger_img_rgb)
```

```python
edge_tiger_img_rgb=cv2.Canny(image=blurred_tiger_img, threshold1=60 , threshold2=100)
plt.imshow(edge_tiger_img_rgb,cmap='gray')
```

```python
edge_tiger_img_rgb=cv2.Canny(image=blurred_tiger_img, threshold1=50 , threshold2=110)
plt.imshow(edge_tiger_img_rgb,cmap='gray')
```

# face detection haar cascade

```python
face_image = cv2.imread('selfie.jpg')
face_image=cv2.cvtColor(face_image,cv2.COLOR_BGR2RGB)
plt.imshow(face_image)
```

```python
haar_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face = haar_cascade.detectMultiScale(face_image) 
face
```

```python
face[0][0]
```

```python
left_top_x=face[0][0]
left_top_y=face[0][1]
width = face[0][2]
height = face[0][3]
```

```python
box = cv2.rectangle(face_image, (left_top_x,left_top_y), (left_top_x+width,left_top_y+height), (0,255,0), 20)
plt.imshow(box)
```

```python
#face_alone = face_image[186:186+332,400:400+332]
face_alone = face_image[left_top_y:left_top_y+height,left_top_x:left_top_x+width]
plt.imshow(face_alone)
```

```python
blurred_face=cv2.blur(face_alone,(31,31))
plt.imshow(blurred_face)
```

```python
#box[186:186+332,400:400+332] =blurred_face
box[left_top_y:left_top_y+height,left_top_x:left_top_x+width]=blurred_face
plt.imshow(box)
```
